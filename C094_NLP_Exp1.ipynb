```python
import pandas as pd
import nltk
import spacy
```

```python
df = pd.read_csv('Test.csv')
df = df.text
df
```

```python
# Punctuational Removal
import string
df_punc = df.str.translate(str.maketrans("","", string.punctuation))
df_punc
```

```python
# Lowering the text
df_lower = df_punc.str.lower()
df_lower
```

```python
import re

# substituting digit by space
df_digit = re.sub(r'\d+','',str(df_lower))
print(df_digit)
```

```python
# initialize stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words_nltk = set(stopwords.words('english'))
print(stop_words_nltk)
```

```python
from gensim.parsing.preprocessing import remove_stopwords
df_without_stop = remove_stopwords(df_digit)
print(df_without_stop)
```

```python
# Tokenization
from nltk.tokenize import word_tokenize
nltk.download('punkt')
tokenized_df = word_tokenize(df_without_stop)
print(tokenized_df)
```

```python
# Stemming
from pprint import pprint
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer= PorterStemmer()
print("Before Stemming")
print(df_without_stop)
print("After Stemming")
for word in tokenized_df:
  print(stemmer.stem(word),end=" ")
```

```python
# lemetizing the text
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
print("After Lemetization")
lemm = WordNetLemmatizer()
for word in tokenized_df:
  print(lemm.lemmatize(word),end=" ")
```

1. <b>What is the advantage of normalizing a word?</b><br>
Ans: In the context of natural language processing (NLP), normalizing words is a common preprocessing step that can help improve the effectiveness of NLP algorithms. Normalization can take many forms, but some common techniques include:<br><br>
  - Case normalization: This involves converting all words to a standard case, such as lowercase or uppercase. This can be useful for tasks such as text classification, where the case of a word might not be relevant and could otherwise cause confusion.<br><br>
  - Stemming: This involves reducing a word to its base form, known as the stem. For example, the stem of the word "jumping" is "jump". Stemming can be useful for tasks such as information retrieval, where the root form of a word is more important than its inflected form.<br><br>
  - Lemmatization: This involves reducing a word to its base form, known as the lemma. Unlike stemming, lemmatization takes into account the part of speech and inflection of a word, so the resulting lemma is a real word. For example, the lemma of the word "jumping" is "jump". Lemmatization can be useful for tasks such as text classification, where the meaning of a word is important.<br>

By normalizing words, NLP algorithms can more easily process and analyze text data, which can improve their accuracy and efficiency. Normalization can also help to reduce the dimensionality of the data, which can be beneficial for tasks such as feature selection and dimensionality reduction.

2. <b>Using TF-IDF (Term Frequency - Inverse Document Frequency) values for features in a uni-gram bag-of-words model should have an effect most similar to which pre processing technique?</b><br>
Ans: The use of term frequency-inverse document frequency (TF-IDF) values as features in a unigram bag-of-words model is most similar to the preprocessing technique of term weighting.<br><br>
TF-IDF is a measure of the importance of a word in a document or a collection of documents. It is calculated by multiplying the term frequency (TF) of a word by the inverse document frequency (IDF) of the word. The term frequency is the number of times a word appears in a document, and the inverse document frequency is a measure of how rare a word is across a collection of documents.<br><br>
TF-IDF values are often used as features in natural language processing (NLP) and information retrieval tasks, such as text classification and document retrieval. By using these values as features, the model can take into account the importance of each word in the document, rather than just its presence or absence.<br><br>
In a unigram bag-of-words model, each word in a document is treated as a separate feature, and the model uses these features to predict a label or class for the document. By using TF-IDF values as features, the model can weigh the importance of each word in the prediction process. This can be particularly useful in cases where certain words are more indicative of the label or class than others.

3. <b>What is irregular noun Form?</b><br>
Ans: In English grammar, an irregular noun form is a noun that does not follow the regular patterns of noun formation. Nouns in English can be regular or irregular, depending on their spelling and pronunciation.<br>
In the context of natural language processing (NLP), irregular noun forms can pose a challenge for algorithms and systems that rely on the regular patterns of language to analyze and understand text.<br>
For example, a machine learning model that has been trained to identify nouns based on certain suffixes or inflections may struggle to recognize irregular nouns, as they do not follow the expected patterns. This can lead to errors or inconsistencies in the model's performance, particularly when processing texts that contain a large number of irregular nouns.<br>
To address this issue, NLP systems may use techniques such as lemmatization or stemming to convert irregular noun forms to their base form, which can make it easier for the model to recognize and process them. Additionally, NLP systems may be trained on large, diverse datasets that include a wide range of irregular noun forms, in order to improve their ability to handle these forms.<br>
Overall, irregular noun forms can be a challenge for NLP systems, but there are strategies and techniques that can help to mitigate this issue and improve the accuracy and efficiency of the system.
